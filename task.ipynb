# Lab Assignment — Improving a RAG System for a Deep Learning Book

## Objective

In this lab, you will improve a **Retrieval-Augmented Generation (RAG) system** built on top of the book **Deep Learning (Goodfellow, Bengio, Courville)**.

Your goal is to progressively enhance the **retrieval quality** of the system so that answers become:

* More accurate
* Better grounded in the source text
* Supported by the correct citations

The tasks start **simple** so you can quickly see progress, then move toward **more advanced improvements**.

---

# Dataset

The system uses the book:

```
Deep Learning — Ian Goodfellow
```

The book has been converted from **PDF → text → chunks → embeddings → PostgreSQL (pgvector)**.

Your system retrieves the most relevant chunks to answer questions.

Example questions:

```
What is backpropagation?
Explain gradient descent.
What is representation learning?
```

---

# Task 1 — Run the RAG System (Easy)

Run the existing RAG pipeline and ask at least **5 questions** about the book.

Example questions:

```
What is gradient descent?
Explain stochastic gradient descent.
What is representation learning?
What are convolutional networks?
What is backpropagation?
```

### Deliverables

Submit:

* The **question**
* The **generated answer**
* The **retrieved source sections**

Example format:

```
Question:
Explain gradient descent

Answer:
...

Sources:
Chapter: Numerical Computation
Chapter: Optimization for Training Deep Models
```

# Task 2 — Analyze the Retrieval (Easy)

Now inspect the retrieved chunks.

For each question:

* Are the retrieved sections relevant?

* Are there irrelevant chunks?

### Write a short analysis

# Task 5 — Implement Hybrid Search

Currently the system uses **vector search only**.

Add **keyword search** and combine both scores.

Example:

```
final_score =
0.7 * vector_similarity
+ 0.3 * keyword_score
```

### Goal

Improve retrieval for **technical terms** such as:

```
SGD
CNN
Backpropagation
```

# Task 6 — Implement Re-ranking

After retrieving the top **20 chunks**, apply a **cross-encoder reranker**.

Example models:

```
bge-reranker-large
ms-marco-MiniLM
```

Pipeline:

```
vector search → top 20
reranker → top 5
LLM → answer
```

### Goal

Ensure the most relevant chunks appear in the final context.
