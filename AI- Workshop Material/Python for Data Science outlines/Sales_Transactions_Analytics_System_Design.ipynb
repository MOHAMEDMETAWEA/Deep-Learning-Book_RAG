## Sales Transactions Analytics API
---

## 1. Project Overview
You are working as a **Data Scientist / Data Engineer** in an e-commerce company.

The company collects **sales transactions** from different systems (website, mobile app, POS). These systems send data in **JSON format**, but the data is often:
- Nested
- Inconsistent
- Partially missing
- Not ready for analytics or machine learning

Your task is to build a **Sales Transactions Analytics API** that receives raw sales data, organizes it using proper **data containers**, applies safe **error handling**, and produces a clean **analytics summary**.

In this version of the project, you are also required to apply **basic Object-Oriented Programming (OOP)** to organize the system components.

> ⚠️ This project does **NOT** include training a machine learning model.
> The focus is on **data understanding, structure, processing, and system design**.

---

## 2. Learning Objectives
By completing this project, you should be able to:
- Work with complex and nested JSON data
- Choose the correct Python data containers (List, Tuple, Set, Dict)
- Design a multi-stage API data pipeline
- Apply `try / except` for safe error handling
- Organize backend systems using basic OOP principles

---

## 3. Input Data Description

The system receives **one JSON object** containing sales transactions.

### Logical Data Structure (Conceptual)

- JSON Object (Dictionary)
  - `orders` → List
    - each order → Dictionary
      - `order_id`
      - `customer_id`
      - `order_date`
      - `items` → List
        - each item → Dictionary
          - `product_id`
          - `category`
          - `price`
          - `quantity`
      - `payment_method`

### Important Notes
- Some orders may have missing fields
- Some items may contain invalid values
- Do NOT assume the input data is clean

---

## 4. API Requirements

Your API must contain **exactly four endpoints**.

### 4.1 Endpoint 1: Health Check

**Purpose**
- Verify that the server is running

**Input**
- No input

**Output**
- A simple status message

---

### 4.2 Endpoint 2: Receive Raw Sales Data

**Purpose**
- Receive and store raw sales data from external systems

**Input**
- JSON object containing sales orders

**Responsibilities**
- Validate the overall structure (e.g., `orders` exists and is a list)
- Store data without cleaning or modification
- Protect the system using `try / except`

**This stage represents:** Raw Data Ingestion

---

### 4.3 Endpoint 3: Clean and Normalize Orders

**Purpose**
- Clean raw sales data and keep only valid items

**Input**
- No new input (uses stored raw data)

**Cleaning Rules**
An item is considered valid if:
- `price` is a number greater than 0
- `quantity` is a number greater than 0
- `category` exists

Invalid items must be ignored safely.

---

### 4.4 Endpoint 4: Sales Analytics Summary

**Purpose**
- Generate a business-friendly analytics summary

**Input**
- No new input (uses cleaned data only)

**Expected Analytics Output (Conceptual)**
- Total number of orders
- Number of unique customers
- Total revenue
- Average order value
- Available product categories
- Payment methods used

---

## 5. Object-Oriented Design Requirement (Mandatory)

You must organize the project using **basic OOP**, without inheritance.

### 5.1 Required Classes

#### Server Class

**Role**
- Represents the API layer
- Defines endpoints and routes
- Delegates processing to the logic layer

**Responsibilities**
- Initialize the API
- Connect endpoints to logic methods

---

#### SalesProcessor Class (API Logic)

**Role**
- Contains all data processing and analytics logic

**Responsibilities**
- Store raw sales data
- Clean and normalize orders
- Organize data using containers
- Compute analytics summaries
- Handle errors using `try / except`

---

#### Client Class

**Role**
- Simulates an external system consuming the API

**Responsibilities**
- Send JSON requests
- Call API endpoints
- Inspect and print responses

---

## 6. Data Containers Usage (Mandatory)

| Container | Purpose |
|---------|--------|
| List | Orders, items, numeric values |
| Tuple | Fixed ranges and limits |
| Set | Unique customers, categories |
| Dict | Orders, items, summaries |

---

## 7. Error Handling Requirements

You must use `try / except` in critical locations:
- Invalid JSON structure
- Missing or incorrect fields
- Type conversion errors
- Numeric calculation failures

The goal is to manage errors safely, not hide them.
