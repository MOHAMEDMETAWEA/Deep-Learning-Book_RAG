```python
import numpy as np  # np is alias for numpy
import time
npArray =  np.arange(0, 100000000)  # similar to range() but returns numpy array

startTime = time.time()
sqArray = npArray ** 2  # element-wise squaring
endTime = time.time()
print("Time taken using NumPy Array: ", endTime - startTime, " seconds")
```

```python
# age , height, weight , billsales
person1 = [25, 175, 70, 1000]  # list
person2 = [30, 180, 80, 1200]
person3 = [35, 165, 60, 900]
person4 = [40, 170, 75, 1100]
people = [person1, person2, person3, person4]  # 2D list
npPeople = np.array(people)  # convert to numpy array
#print npPeople print before discount
print("Before discount:\n", npPeople)
```

```python
billsales = npPeople[:,3]
discountBill = billsales  - (billsales * 0.10) 
npPeople[:,3] = discountBill
print("After discount:\n", npPeople)
```

```python
# big o notation n^2
for person in people:
    original_bill = person[3]
    discounted_bill = original_bill * 0.9
    person[3] = discounted_bill
print("After discount:\n", npPeople)
```
